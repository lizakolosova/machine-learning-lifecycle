As we saved preprocessing pipeline, to have a full pipeline with a model and everything, we reload it here.

```python
from sklearn.pipeline import Pipeline
import joblib
```

```python
prep_package = joblib.load("artifacts/mushroom_preprocessing_pipeline.pkl")
prep_pipeline = prep_package["pipeline"]

best_model = joblib.load("artifacts/Decision Tree.pkl")

final_pipeline = Pipeline([
    ("preprocessing", prep_pipeline),
    ("model", best_model)
])

joblib.dump(final_pipeline, "artifacts/final_mushroom_pipeline.pkl")
```
